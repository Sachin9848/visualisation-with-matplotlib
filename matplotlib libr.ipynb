```python
#scatter plot
#positive corelation plotting
import matplotlib.pyplot as plt
x_values = [0,1,2,3,4,5]
squares = [0,1,4,9,16,25]

#this s stands for size, and it allows us too control how big we want the points on the graph
plt.scatter(x_values,squares, s=100,color = 'Blue')
plt.show()
```

```python
#negative corelation plotting
x_values = [6,7,8,9,10]
squares = [10,9,8,7,6]

#this s stands for size, and it allows us too control how big we want the points on the graph
plt.scatter(x_values,squares, s=200,color = 'Blue')
plt.show()
```

```python
#pie chart
langs=['c','c++','java','python','PHP']
students=[23,17,35,29,12]
plt.pie(students,labels=langs,autopct='%.2f%%')
plt.title('pie chart')
plt.show()
```

```python
#Autopct string, used to label the wedges with their numeric value. The label will be placed inside the wedge.

#The format string will be fmt%pct
```

```python
my_explode=(0,0.1,0.4,0,0)
langs=['c','c++','java','python','PHP']
students=[23,17,35,29,12]
plt.pie(students,labels=langs,shadow=True,autopct='%.2f%%',

       
        
       explode=my_explode)

plt.title('pie chart')
plt.show()
```
